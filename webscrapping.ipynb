## Selecting the seasons to analyse

```python
# Declarar la lsita con los nombres de las hojas de c√°lclulo
temporadas = [
     # '2012-2013',
     '2013-2014'
    ,'2014-2015'
    ,'2015-2016'
    ,'2016-2017'
    ,'2017-2018'
    ,'2018-2019'
    ,'2019-2020'
    ,'2020-2021'
    ,'2021-2022'
    ,'2022-2023_J15'
    ]
```

## Plotting all the best teams in the seasons

```python
for temporada in temporadas:
    # Cargar el archivo de hoja de c√°lculo
    dataframe = pd.read_excel(path + temporada + ".xlsx")
    dataframe = dataframe.sort_values(by=dataframe.columns[-1], ascending=False)
    dataframe = dataframe.reset_index(drop = True)

    dataframe.set_index('name', append=True, inplace=True)
    jornadas = dataframe.columns

    fig, ax = plt.subplots(figsize=(20, 5))
    for name in dataframe.index.get_level_values('name')[0:6]:
        plt.title("TEMPORADA " + temporada)
        ax.plot([x+1 for x in list(range(0, len(jornadas)))], dataframe[dataframe.index.get_level_values('name') == name].values.reshape(-1,1))
        plt.legend(dataframe.index, loc ="upper left")
        plt.xticks([x+1 for x in list(range(0, len(jornadas)))])
```
